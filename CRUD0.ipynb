```python
# Student Management System - CLI Based (CRUD)

import os
import json

STUDENT_FILE = 'students.json'

def load_students():
    if os.path.exists(STUDENT_FILE):
        with open(STUDENT_FILE, 'r') as f:
            return json.load(f)
    return {}

def save_students(data):
    with open(STUDENT_FILE, 'w') as f:
        json.dump(data, f, indent=4)

def create_student():
    students = load_students()
    roll = input("enter roll number: ")
    if roll in students:
        print("Student already exists.")
        return
    name = input("Enter name:")
    age = input("Enter Age:")
    students[roll] = {'name': name, 'age': age, 'grade': input("Enter Grade:A ")}
    save_students(students)
    print("Student added successfully.")

def read_students():
    students = load_students()
    if not students:
        print("No student records found.")
    else:
        for roll, info in students.items():
            print(f"Roll: {roll}, Name: {info['name']}, Age: {info['age']}, Grade: {info['grade']}")

def update_student():
    students = load_students()
    roll = input("age ")
    if roll not in students:
        print("Student not found.")
        return
    name = input("Enter New Name: ")
    age = input("Enter New Age: ")
    grade = input("Enter New Grade: ")
    students[roll] = {'name': name, 'age': age, 'grade': grade}
    save_students(students)
    print("Student record updated.")

def delete_student():
    students = load_students()
    roll = input("Enter Roll Number to delete: ")
    if roll in students:
        del students[roll]
        save_students(students)
        print("Student record deleted.")
    else:
        print("Student not found.")

def search_student():
    students = load_students()
    search_term = input("Enter Name or Roll Number to search: ").lower()
    found = False
    for roll, info in students.items():
        if search_term in roll.lower() or search_term in info['name'].lower():
            print(f"Found - Roll: {roll}, Name: {info['name']}, Age: {info['age']}, Grade: {info['grade']}")
            found = True
    if not found:
        print("No matching student found.")

def menu():
    while True:
        print("\n--- Student Management System ---")
        print("1. Add Student")
        print("2. View Students")
        print("3. Update Student")
        print("4. Delete Student")
        print("5. Search Student")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            create_student()
        elif choice == '2':
            read_students()
        elif choice == '3':
            update_student()
        elif choice == '4':
            delete_student()
        elif choice == '5':
            search_student()
        elif choice == '6':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
menu()
```
